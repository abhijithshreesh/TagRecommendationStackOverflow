```python
import pandas as pd
import numpy as np
```

```python
X_train = pd.read_csv("data/X_Train.csv")
Y_train = pd.read_csv("data/Y_Train.csv")
```

```python
X_test = pd.read_csv("data/X_Test.csv")
```

```python
from skmultilearn.problem_transform import BinaryRelevance
from sklearn import tree
```

```python
classifier = BinaryRelevance(classifier = tree.DecisionTreeClassifier(), require_dense = [False, True])

# train
classifier.fit(X_train.values, Y_train.values)
```

```python
# predict
predictions = classifier.predict(X_test.values)
predictions.count_nonzero()
```

```python
np.savetxt("data/dtree_test.csv", predictions.toarray(), delimiter=",")
```

```python
total_tags = pd.read_csv("data/allTags.csv")
```

```python
total_tags = total_tags.values.T[0]
```

```python
final_result = []
for prediction in predictions.toarray():
    result = ""
    for elem in range(len(prediction)):
        if np.asscalar(prediction[elem]) is 1:
            result = result + total_tags[elem] + ","
    final_result.append(result)
```

```python
final_predictions = pd.DataFrame(np.asarray(final_result))
```

```python
total_df = pd.read_csv("data/cleaned_data.csv", encoding = "ISO-8859-1")
total_df = total_df[~total_df.Tags.isnull()]
total_records = len(total_df)
test_df = total_df.head(int(total_records * .23))
```

```python
final_result = [test_df['stemmed_words'], test_df['Tags'], final_predictions]
final_result = pd.concat(final_result, axis=1)
```
